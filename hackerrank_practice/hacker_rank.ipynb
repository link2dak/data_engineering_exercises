# stack cubes

```python
t = int(input())

for _ in range(t):
    n = int(input())
    cubes = list(map(int, input().split()))

    left = 0
    right = n - 1
    last = float("inf")

    possible = True

    while left <= right:

        # Choose the larger end
        if cubes[left] > cubes[right]:
            current = cubes[left]
            left += 1
        else:
            current = cubes[right]
            right -= 1

        # Does it fit on the stack?
        if current > last:
            possible = False
            break

        last = current

    if possible:
        print("Yes")
    else:
        print("No")
```

# duplicate records while preserving order

```python
def count(li:list):
    new_lis = []
    for i in li:
        if i not in new_lis:
            new_lis.append(i)
    return new_lis

print(count([
    "user1",
    "user2",
    "user1",
    "user3",
    "user2",
    "user4"
]))
```

# ginorts

```python
s = "Sorting1234"


lowercase = []
uppercase = []
odd = []
even = []
for i in s:
    if i.isupper():
        uppercase.append(i)
    elif i.islower():
        lowercase.append(i)
    elif int(i) % 2 == 0:
        even.append(i)
    else:
        odd.append(i)

lowercase.sort()
uppercase.sort()
odd.sort()
even.sort()

merge = lowercase + uppercase + odd + even
final_s = ""
for i in merge:
    final_s+=i
print(final_s)
```
