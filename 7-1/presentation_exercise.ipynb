```python
from pyspark import SparkContext

sc = SparkContext("local[*]", "PartitionExample")

# Create key-value RDD (required for partitionBy)
data = sc.parallelize([
    ("A", 1), ("B", 2), ("C", 3),
    ("A", 4), ("B", 5), ("C", 6)
])

# 1. Partition by key into 4 partitions
partitioned = data.partitionBy(3)

print("After partitionBy:", partitioned.getNumPartitions())
print("Partition contents:", partitioned.glom().collect())

# 2. Reduce partitions using coalesce
reduced = partitioned.coalesce(1)

print("After coalesce:", reduced.getNumPartitions())
print("After coalesce contents:", reduced.glom().collect())

sc.stop()
```
