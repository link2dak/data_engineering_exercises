# This is the friends by age challenge with dataframes

```python
from pyspark.sql import SparkSession

spark = SparkSession.builder.appName("FakeFriends").getOrCreate()

# Load CSV
df = spark.read.csv("../../fakefriends-header.csv", header=True, inferSchema=True)

# Select name and age
result = df.select("name", "age")

# Print all rows
result.show()
```
