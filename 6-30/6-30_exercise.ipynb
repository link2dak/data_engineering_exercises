# This is the customer sorted challenge

```python
from pyspark import SparkConf, SparkContext
from prettytable import PrettyTable
import re

conf = SparkConf().setMaster('local[*]').setAppName("wordCount")
sc = SparkContext(conf = conf)

customer = sc.textFile("../../customer-orders.csv")

# get only the values of the customer id and how much they spent
tuple = customer.map(lambda line: (line.split(",")[0], float(line.split(",")[2])))
# group by the id and add all the totals together
compressed = tuple.reduceByKey(lambda x, y: x+y)

# sory by the total amount in descending order
sorted = compressed.sortBy(lambda x: x[1], ascending=False)


table = PrettyTable()
table.field_names = ["ID", "Total"]

for id, total in sorted.collect():
    table.add_row([id, f"{total:.2f}"])
print(table)
sc.stop()
```
